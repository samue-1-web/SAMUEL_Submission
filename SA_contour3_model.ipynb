```python
import torch
import torch.nn as nn

N_POINTS = 100

class ResidualContourModel(nn.Module):

    def __init__(self):
        super().__init__()

        self.network = nn.Sequential(

            nn.Linear(400, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(),
            nn.Dropout(0.2),

            nn.Linear(256, 128),
            nn.BatchNorm1d(128),
            nn.ReLU(),
            nn.Dropout(0.2),

            nn.Linear(128, 200)

        )

        nn.init.zeros_(self.network[-1].weight)
        nn.init.zeros_(self.network[-1].bias)

    def forward(self, prev_pts, future_pts):

        batch = prev_pts.size(0)

        x = torch.cat(
            [
                prev_pts.view(batch, -1),
                future_pts.view(batch, -1)
            ],
            dim=1
        )

        return self.network(x).view(batch, N_POINTS, 2)
```
